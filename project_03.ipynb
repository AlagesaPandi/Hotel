```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", None)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("hotel_bookings_updated_2024.csv")
```

```python
print("Shape of Dataset:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nDataset Information:")
df.info()
```

```python
df.isnull().sum()
```

```python
df["children"] = df["children"].fillna(0)

if "country" in df.columns:
    df["country"] = df["country"].fillna(df["country"].mode()[0])

if "agent" in df.columns:
    df["agent"] = df["agent"].fillna(0)

if "company" in df.columns:
    df["company"] = df["company"].fillna(0)

print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x="is_canceled", data=df)

plt.title("Booking Cancellation")
plt.xlabel("Cancelled (0=No,1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(y="hotel", data=df)

plt.title("Hotel Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="customer_type", data=df)

plt.xticks(rotation=20)
plt.title("Customer Type")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="meal", data=df)

plt.title("Meal Preference")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.countplot(y="market_segment", data=df)

plt.title("Market Segment")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(y="distribution_channel", data=df)

plt.title("Distribution Channel")
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.countplot(x="arrival_date_month", data=df)

plt.xticks(rotation=45)
plt.title("Monthly Bookings")
plt.show()
```

```python
plt.figure(figsize=(10,5))

df["city"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Booking Cities")
plt.xlabel("City")
plt.ylabel("Bookings")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x="reservation_status", data=df)

plt.title("Reservation Status")
plt.xticks(rotation=20)
plt.show()
```
